```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
from numpy.random import randn
```

```python
ser1=Series([1,2,3,4], index=['a', 'b', 'c', 'd'])
ser1
```

```python
ser2=ser1.reindex(['a', 'b', 'c', 'd', 'e', 'f'])
ser2
```

```python
ser2.reindex(['a', 'b', 'c', 'd', 'e', 'f','g'],fill_value=0)
```
