```python
dframe=DataFrame({'k1':['X', 'Y', 'Y', 'Z'], 
                  'k2':['alpha', 'beta', 'alpha', 'beta'],'dataset1':np.random.randn(4),'dataset2':np.random.randn(4)})
```

```python
group1=dframe['dataset1'].groupby(dframe['k1'])
```
